```python
import re
```

```python

fifa['Contract_Start'] = fifa['Contract'].apply(lambda x: re.search(r'\d{4}', x).group() if re.search(r'\d{4} ~ \d{4}', x) else pd.NA)
fifa['Contract_end'] = fifa['Contract'].apply(lambda x: re.search(r'~ (\d{4})', x).group(1) if re.search(r'\d{4} ~ \d{4}', x) else pd.NA)

fifa.drop(columns=['Contract'],inplace=True)
fifa['Contract_Start']=pd.to_numeric(fifa['Contract_Start'].str.strip())
fifa['Contract_end']=pd.to_numeric(fifa['Contract_end'].str.strip())
```

```python
fifa.drop(columns='Positions',inplace=True)
```

```python
from datetime import datetime

def date_obj(x):
    date = datetime.strptime(x, "%b %d, %Y")
    return date.strftime("%d-%m-%Y")


```

```python
fifa['Joined']=fifa['Joined'].apply(date_obj)
```

```python
fifa['Loan Date End'].fillna('null',inplace=True)
```

```python
fifa['Weight'].unique()
```

```python
def height_to_cm(x):
    try:
        if 'cm' in x:
            return int(x.replace('cm', ''))
        elif '\'' in x:
            feet, inches = map(int, x.replace('\'', '').split('\''))
            return (feet * 30.48) + (inches * 2.54)
        else:
            return None  
    except ValueError:
        return None  


fifa['Height(cm)'] = fifa['Height'].apply(height_to_cm)
```

```python
def wg_kg(x):
    if 'kg' in x:
        return int(x.replace('kg',''))
    elif 'lbs' in x:
        r=int(x.replace('lbs',''))
        return r*0.453592
    else:
        return None
```

```python
fifa['Weight(Kg)'] = fifa['Weight'].apply(wg_kg)
```

```python
fifa.drop(columns=['Height','Weight'],inplace=True)
```

```python
def to_m(x):
    if 'M' in x:
        return float(x.replace('€','').replace('M',''))*1000000
    elif 'K' in x:
        return float(x.replace('€','').replace('K',''))*1000
```

```python
fifa['Value']=fifa['Value'].apply(to_m)
fifa['Wage']=fifa['Wage'].apply(to_m)
```

```python
fifa['Release Clause']=fifa['Release Clause'].apply(to_m)
```

```python
def no_star(x):
    if '★' in x:
        return int(x.replace('★',''))
```

```python
fifa['Weak_foot']=fifa['W/F'].apply(no_star)
```

```python
fifa['Skill_moves']=fifa['SM'].apply(no_star)
```

```python
fifa['international_reputation']=fifa['IR'].apply(no_star)
```

```python
fifa.drop(columns=['W/F','SM','IR','LongName'],inplace=True)
```

```python
import numpy as np  
def to_k(x):
    if isinstance(x, str):
        if 'M' in x:
            return int(float(x.replace('M', '')) * 1000000)
        elif 'K' in x:
            return int(float(x.replace('K', '')) * 1000)
        else:
            return int(x) 
    elif not np.isnan(x):
        return int(x) 
    else:
        return x 

```

```python
fifa['Hits'] = fifa['Hits'].apply(to_k)
```
